```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pulp
from haversine import haversine, Unit
```

```python
df = pd.read_csv('gifts.csv')
df = df.sample(n=500, replace=False)
northpole = pd.DataFrame([[0, 90, 0, 0]], columns=['GiftId', 'Latitude', 'Longitude', 'Weight']).iloc[0]
```

```python
sorted_weight = df.sort_values(by=['Weight'],ascending=False)
sorted_weight
```

```python
x = sorted_weight["Weight"]
plt.style.use('ggplot')
plt.hist(x, bins=20)
plt.show()
```

```python
def findNext(startPoint,points):
    bestDistance = float('inf')
    distances = np.zeros(len(points))
    indices = np.zeros(len(points)).astype(int)
    i = 0
    for index, point in points.iterrows():
        distances[i] = haversine((startPoint['Latitude'],startPoint['Longitude']),(point['Latitude'],point['Longitude']))
        indices[i] = index
        i += 1
    best_ind = np.argmin(distances)
    return points.iloc[best_ind]
```

```python
currentpoint = northpole
remainingVectors = df
path = []

while len(remainingVectors) > 0:
    currentpoint = findNext(currentpoint,remainingVectors)
    remainingVectors = remainingVectors[remainingVectors.GiftId != currentpoint[0]]
    path.append(currentpoint)


path
```
